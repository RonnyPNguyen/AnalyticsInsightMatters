```python
import pandas as pd
import os
```

```python
df = pd.read_excel('dataJune.xlsx', sheet_name='data')
df['dateAdded']= pd.to_datetime(df['dateAdded'], format='%d/%m/%Y')
# Format specified columns to numeric type
cols_to_format = [
    "askingPrice", "salePrice", "annualRevenue", "annualOutgoings", "annualWages", "annualRent", "annualProfit", "gstIncluded",
    "savIncluded", "yieldIncluded", "yieldValue", "sgaIncluded", "sgaValue", "tradingDays", "listingStatus", "listingVerified"]
df[cols_to_format] = df[cols_to_format].apply(pd.to_numeric, errors='coerce')
# Convert all columns except those already formatted to string
cols_to_exclude = cols_to_format + ['dateAdded']
cols_to_convert = [col for col in df.columns if col not in cols_to_exclude]
df[cols_to_convert] = df[cols_to_convert].astype(str)

df
```

```python
df.to_json('../backend/data/listingsData.json', orient='records', date_format='iso', force_ascii=False, indent=4)
```
